```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
#Lendo o arquivo csv de precipitacao
df_precipitacao = pd.read_csv('inmet/all_cities.csv', sep=',')
```

```python
#Criando dataframe de precipitacao por ano/mes de cada estacao
precipitacao_mes_ano = pd.DataFrame({'Precipitacao' : df_precipitacao.groupby(["Estacao", "year","month"])["Precipitacao"].sum()}).reset_index()
```

```python
#Alterando os nomes das colunas do dataframe precipitacao_mes_ano
precipitacao_mes_ano.columns = ['ESTACAO','ANO','MES','PRECIPITACAO']
```

```python
precipitacao_mes_ano_2010_2012 = precipitacao_mes_ano.loc[(precipitacao_mes_ano['ANO'] < 2013) & (precipitacao_mes_ano['ANO'] > 2009)]
```

```python
"""
PRECIPITACAO DE TODAS AS ESTACOES POR ANO DE 2000 a 2012
"""

#Criando dataframe de precipitacao por ano de cada estacao
precipitacao_ano = pd.DataFrame({'PRECIPITACAO' : precipitacao_mes_ano.groupby(["ESTACAO","ANO"])["PRECIPITACAO"].sum()}).reset_index()
```

```python
#Precipitacao de todas as estacoes por ano de 2010 a 2012 
precipitacao_ano_2010_2012 = precipitacao_ano.loc[(precipitacao_ano['ANO'] < 2013) & (precipitacao_ano['ANO'] > 2009)]
```

```python
# ---------- RECIFE ----------
#Dataframe Precipitacao estacao de Recife todos os anos
df_recife = precipitacao_ano.loc[precipitacao_ano['ESTACAO'] == 'RECIFE']
```

```python
#Grafico Precipitacao de Recife todos anos
df_recife.plot(x='ANO',y='PRECIPITACAO',kind='scatter', title='Precipitacao de Recife 2000/2012')
```

```python
#Descobrindo o ano com maior precipitacao em Recife entre 2010 e 2012
df_recife_2010_2012 = precipitacao_ano_2010_2012.loc[(precipitacao_ano_2010_2012['ESTACAO'] == 'RECIFE')]
ano_recife_maior_precipitacao = df_recife_2010_2012[df_recife_2010_2012['PRECIPITACAO']==df_recife_2010_2012['PRECIPITACAO'].max()]['ANO'].values[0]
```

```python
#Criando dataframe de precipitacao de Recife no ano de maior precipitacao mes a mes
recife_maior_precipitacao_2010_2012 = precipitacao_mes_ano_2010_2012.loc[(precipitacao_mes_ano_2010_2012['ESTACAO'] == 'RECIFE') & (precipitacao_mes_ano_2010_2012['ANO'] == ano_recife_maior_precipitacao)]
```

```python
#Grafico de Precipitacao do ano com maior precipitacao do Recife entre 2010 e 2012
recife_maior_precipitacao_2010_2012.plot(x='MES',y='PRECIPITACAO',kind='scatter', title='Precipitacao de Recife ('+str(ano_recife_maior_precipitacao)+')')
```

```python
#Aplicando correlacao Spearman - nativo do Pandas
recife_maior_precipitacao_2010_2012['MES'].corr(recife_maior_precipitacao_2010_2012['PRECIPITACAO'], 'spearman')
```
